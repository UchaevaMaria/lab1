```python
import matplotlib.pyplot as plt
import numpy as np

#   
x = np.linspace(0, 10, 100)  #100   0  10
y = np.sin(x)  #   
z = np.cos(x)
#  
plt.plot(x, y, label='sin(x)', color='green')
plt.plot(x, z, label='cos(x)', color='blue')
#    
plt.title('  ')
plt.xlabel('x')
plt.ylabel('sin(x) cos(x)')
plt.legend()
plt.show()
```

```python
import numpy as np
matrix = np.random.randint(1, 11, size=(5, 5))
print(" :\n", matrix)
average = np.mean(matrix)
print()
print(f" : {average}")
max_m = np.max(matrix)
print(f"  {max_m}")
min_m = np.min(matrix)
print(f" :{min_m}")
column_sums = np.sum(matrix, axis=0)
print(f"   {column_sums}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
matrix = np.array([[8, 1, 2, 9, 5],
                   [2, 6, 5, 10, 9],
                   [9, 10, 3, 4, 3],
                   [2, 7, 9, 9, 6],
                   [9, 2, 10, 8, 5]]) #   
fig, ax = plt.subplots()
im = ax.imshow(matrix) #   
cbar = ax.figure.colorbar(im, ax=ax) #   
plt.title("  (Heatmap)")
ax.set_xlabel("", color = 'red')
ax.set_ylabel('', color='red')
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        text = ax.text(j, i, matrix[i, j], color="w")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
x = np.linspace(-5, 5, 100)  #   x  -5  5  100 
y = x**2
ax[0].plot(x, y, color = '#eb4b19')
ax[0].set_title("  y = x^2")
ax[0].set_facecolor("lightgrey")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
num_points = 50
x_scatter = np.random.rand(num_points)
y_scatter = np.random.rand(num_points)
ax[1].scatter(x_scatter, y_scatter, color ='#9744c1')
ax[1].set_title(' ')
ax[1].set_facecolor("#b2d7bb")
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")
categories = ['A', 'B', 'C']
values = [3, 7, 2]
ax[2].bar(categories, values, color = 'pink')
ax[2].set_title(" ", color = '#6f2f50')
ax[2].set_xlabel("")
ax[2].set_ylabel("")

plt.show()
```
