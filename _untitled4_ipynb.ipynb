```python
import matplotlib.pyplot as plt
import numpy as np

#создаем данные для графика
x = np.linspace(0, 10, 100)  #100 точек от 0 до 10
y = np.sin(x)  # значения функции синуса
z = np.cos(x)
# Создаем график
plt.plot(x, y, label='sin(x)', color='green')
plt.plot(x, z, label='cos(x)', color='blue')
#добавляем заголовок и метки осей
plt.title('график функции синуса')
plt.xlabel('x')
plt.ylabel('sin(x) cos(x)')
plt.legend()
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  # 1000 случайных чисел из нормального распределения

#построение гистограммы
plt.hist(data, bins=20, alpha=0.3, color='blue')
plt.title('Гистограмма случайных данных')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.grid(axis='y', alpha=0.75)
plt.grid(axis='x', alpha=0.75) #создание сеток
plt.show()
```

```python
import numpy as np
matrix = np.random.randint(1, 11, size=(5, 5))
print("исходная матрица:\n", matrix)
average = np.mean(matrix)
print()
print(f"среднее значение: {average}")
max_m = np.max(matrix)
print(f"максимальное значение {max_m}")
min_m = np.min(matrix)
print(f"минимальное значение:{min_m}")
column_sums = np.sum(matrix, axis=0)
print(f"суммма по столбцам {column_sums}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
matrix = np.array([[8, 1, 2, 9, 5],
                   [2, 6, 5, 10, 9],
                   [9, 10, 3, 4, 3],
                   [2, 7, 9, 9, 6],
                   [9, 2, 10, 8, 5]]) #матрица из прошлого задания
fig, ax = plt.subplots()
im = ax.imshow(matrix) #отобр изобр на граф
cbar = ax.figure.colorbar(im, ax=ax) #добавляем цветовую полосу справа
plt.title("тепловая карта (Heatmap)")
ax.set_xlabel("столбцы", color = 'red')
ax.set_ylabel('строки', color='red')
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        text = ax.text(j, i, matrix[i, j], color="w")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
x = np.linspace(-5, 5, 100)  # Создаем массив x от -5 до 5 с 100 точками
y = x**2
ax[0].plot(x, y, color = '#eb4b19')
ax[0].set_title("Линейный график y = x^2")
ax[0].set_facecolor("lightgrey")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
num_points = 50
x_scatter = np.random.rand(num_points)
y_scatter = np.random.rand(num_points)
ax[1].scatter(x_scatter, y_scatter, color ='#9744c1')
ax[1].set_title('точечный график')
ax[1].set_facecolor("#b2d7bb")
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")
categories = ['A', 'B', 'C']
values = [3, 7, 2]
ax[2].bar(categories, values, color = 'pink')
ax[2].set_title("столбчатая диаграмма", color = '#6f2f50')
ax[2].set_xlabel("категории")
ax[2].set_ylabel("значения")

plt.show()
```
